```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sbn
%matplotlib inline
```

```python
columns = ['sepal length' , 'sepal width' , 'petal length' , 'petal width' , 'Class_labels']
#Load the data
df = pd.read_csv('dataset.csv' , names = columns)
df.head(150)
```

```python
df.describe()
```

```python
# Convert relevant columns to numeric type
for col in ['sepal length', 'sepal width', 'petal length', 'petal width']:
    df[col] = pd.to_numeric(df[col], errors='coerce')  # 'coerce' handles non-numeric values

#visualize the whole dataset
sbn.pairplot(df, hue='Class_labels')
```

 **Separating input & output columns**

```python
data = df.values
X = data[:,0:4]
Y = data[:,4]
print(X)
```

Model 1 : Support vector machine algorithm


```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.impute import SimpleImputer

# Impute missing values using SimpleImputer
imputer = SimpleImputer(strategy='mean')  # Replace NaN with the mean of the column
X = imputer.fit_transform(X)

# Now split the data after imputation
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)

# Create and fit the SVC model
model_svc = SVC()
model_svc.fit(X_train, Y_train)
```

```python
prediction1 = model_svc.predict(X_test)
#calculate the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test, prediction1)*100)
for i in range(len(prediction1)):
  print(Y_test[i],prediction1[i])
```
